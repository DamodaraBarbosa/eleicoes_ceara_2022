# Análises gerais sobre as eleições no Ceará

## Quanto ao governo do estado

```python
# os candidatos foram:

candidatos_governador = eleicoes_ce.query('DS_CARGO == "GOVERNADOR"')[['NR_VOTAVEL','NM_VOTAVEL']].drop_duplicates(
    subset= 'NM_VOTAVEL'
).drop([3075, 25537])
candidatos_governador.index = candidatos_governador['NR_VOTAVEL']
candidatos_governador.index.name = 'Número do candidato'
candidatos_governador.pop('NR_VOTAVEL')
candidatos_governador.columns = ['Nome do candidato a governador']
```

```python
candidatos_governador
```

```python
# votos referentes ao governo do estado:

votos_governador = eleicoes_ce.query('DS_CARGO == "GOVERNADOR"').groupby('NM_VOTAVEL')['QT_VOTOS'].sum()
votos_governador.pop('VOTO BRANCO')
votos_governador.pop('VOTO NULO')
```

```python
# gráfico em formato de barras:

votos_governador = votos_governador[
    ['ELMANO DE FREITAS DA COSTA', 'WAGNER SOUSA GOMES', 'ROBERTO CLÁUDIO RODRIGUES BEZERRA']
]
fig = px.bar(x= votos_governador.keys(), y= votos_governador.values, color = votos_governador.keys(),
             color_discrete_map= {'ELMANO DE FREITAS DA COSTA': 'red', 'WAGNER SOUSA GOMES': 'blue', 'ROBERTO CLÁUDIO RODRIGUES BEZERRA': 'green'}
)
fig.update_layout(title= 'Votos por candidato ao governo do Ceará', title_x= 0.5, xaxis_title= 'Candidato',
                  yaxis_title= 'Quantidade de votos', legend_title_text= 'Candidato'
)
fig.show()
```

```python
# gráfico em formato de pizza:

fig = go.Figure()
fig.add_pie(labels= votos_governador.keys() ,values= votos_governador.values)
fig.update_layout(title= 'Votos por candidato ao governo do Ceará', title_x= 0.5)
fig.show()
```

```python
# cidades com maior quantidade de eleitores:
# determina-se a quantidade de eleitores em cada cidade a partir dos votos para governador, definem-se as 10 cidades do estado com maior eleitorado

eleitorado_cidade = eleicoes_ce.query('DS_CARGO == "GOVERNADOR"').groupby(['NM_MUNICIPIO'])['QT_VOTOS'].sum()
top10_cidades = list(eleitorado_cidade.nlargest(10).keys())
```

```python
# quantidade de votos para os 3 principais candidatos a governador em cada cidade:

candidatos = ['WAGNER SOUSA GOMES', 'ELMANO DE FREITAS DA COSTA', 'ROBERTO CLÁUDIO RODRIGUES BEZERRA']
voto_municipio_governador = eleicoes_ce.query('DS_CARGO == "GOVERNADOR" and NM_MUNICIPIO in @top10_cidades and NM_VOTAVEL in @candidatos').groupby(
    ['NM_MUNICIPIO', 'NM_VOTAVEL'])['QT_VOTOS'].sum()

voto_municipio_governador = pd.DataFrame(voto_municipio_governador)
voto_municipio_governador.reset_index('NM_VOTAVEL', inplace= True)
```

```python
# plotando o gráfico:

fig = px.histogram(voto_municipio_governador, x= voto_municipio_governador.index, y= 'QT_VOTOS', color= 'NM_VOTAVEL', barmode= 'group', histfunc= 'avg')
fig.update_layout(title= 'Voto para governador nos municípios com maior eleitorado do Ceará', title_x = 0.5, xaxis_title= 'Municípios',
                  yaxis_title= 'Quantidade de votos', legend_title_text= 'Candidatos'
                  )
fig.show()
```

```python
# candidato "vencedor" em cada cidade:

group_cidade_governador = eleicoes_ce.query('DS_CARGO == "GOVERNADOR"').groupby(['NM_MUNICIPIO', 'NM_VOTAVEL'])['QT_VOTOS'].sum()
group_cidade_governador = pd.DataFrame(group_cidade_governador)
```

```python
group_cidade_governador.reset_index(inplace= True)
```

```python
municipio_dict = dict()
municipios = list()

for municipio in group_cidade_governador['NM_MUNICIPIO'].unique():
  dados_cidade = group_cidade_governador.query(f'NM_MUNICIPIO == "{municipio}"')
  mais_votado_cidade = dados_cidade[dados_cidade['QT_VOTOS'] == dados_cidade['QT_VOTOS'].max()]
  votos_na_cidade = dados_cidade["QT_VOTOS"].sum()
  municipio_dict['municipio'] = mais_votado_cidade['NM_MUNICIPIO'].values[0]
  municipio_dict['candidatovencedor'] = mais_votado_cidade['NM_VOTAVEL'].values[0] 
  municipio_dict['votos'] = mais_votado_cidade['QT_VOTOS'].values[0]
  municipio_dict['porcentagem'] = round(((mais_votado_cidade['QT_VOTOS'].values[0]/votos_na_cidade) * 100), 2)
  municipios.append(municipio_dict.copy())
```

```python
vencedores_por_cidade = pd.DataFrame(municipios)
vencedores_por_cidade
```

```python
vencedores_por_cidade.value_counts('candidatovencedor')

# o candidato Elmano de Freitas foi o mais votado em 178 dos 184 municípios cearenses.
```

```python
# as cidades em que o cadidato Wagner Gomes foi o mais votado foram:

vencedores_por_cidade.query('candidatovencedor == "WAGNER SOUSA GOMES"')
```

```python
# e o candidato Roberto Cláudio foi o mais votado apenas em:

vencedores_por_cidade.query('candidatovencedor == "ROBERTO CLÁUDIO RODRIGUES BEZERRA"')
```

```python
# as 10 cidades com vitórias mais largas foram:

maiores_porcentagens = vencedores_por_cidade['porcentagem'].nlargest(10)
vencedores_por_cidade.query('porcentagem in @maiores_porcentagens')
```

```python
# já na região metropolitana de Fortaleza:

regiao_metropolitana = ['AQUIRAZ', 'CASCAVEL', 'CAUCAIA', 'CHOROZINHO', 'EUSÉBIO', 'FORTALEZA', 'GUAIÚBA', 'HORIZONTE', 'ITAITINGA', 'MARACANAÚ', 'MARANGUAPE',
                        'PACAJUS', 'PACATUBA', 'PINDORETAMA', 'SÃO GONÇALO DO AMARANTE', 'SÃO LUIS DO CURU', 'PARAIPABA', 'PARACURU', 'TRAIRI']

votos_regiao_metropolitana = eleicoes_ce.query('NM_MUNICIPIO in @regiao_metropolitana and DS_CARGO == "GOVERNADOR" and NM_VOTAVEL in @candidatos'
).groupby(['NM_MUNICIPIO', 'NM_VOTAVEL'])['QT_VOTOS'].sum()
votos_regiao_metropolitana = pd.DataFrame(votos_regiao_metropolitana)
votos_regiao_metropolitana.reset_index('NM_VOTAVEL', inplace= True)
```

```python
fig = px.histogram(votos_regiao_metropolitana, x= votos_regiao_metropolitana.index, y= votos_regiao_metropolitana['QT_VOTOS'],
                   color= votos_regiao_metropolitana['NM_VOTAVEL'], barmode= 'group')
fig.update_layout(title= 'Voto por candidato nos municípios da região metropolitana de Fortaleza', title_x= 0.5, xaxis_title= 'Município',
                   yaxis_title= 'Quantidade de votos', legend_title_text= 'Candidatos')
fig.show()
```

```python
#vencedor por cidade da região metropolitana de Fortaleza:

vencedor_reg_metropolitana = vencedores_por_cidade.query('municipio in @regiao_metropolitana')
vencedor_reg_metropolitana
```
